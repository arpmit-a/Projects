```python

import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/water_potability.csv')
df
```

```python
df=df.dropna()
#Dropping off the rows with NaN values
```

```python
df = df.reset_index()
df
```

```python
df.info()
```

```python
x = df.iloc[:,0:9].values #Allocating inputs and outputs with values
y = df.iloc[:,9].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)#Classifying the inputs and outputs as train and test variables
```

```python
from sklearn.svm import SVC
model = SVC() #importing support vector classifier algorithm
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
